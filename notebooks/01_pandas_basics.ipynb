```python
import pandas as pd
df = pd.read_csv('../data/titanic.csv')

df.head()
df.info() #data audit
df.describe() #check for numerical summary statistics

df[df['Fare']==0].shape
df[df['Cabin'].isnull()] #check for zero fare entries

df.loc[0:9, ['Name', 'Age', 'Fare']] #view first 10 rows of selected columns
df.iloc[-1]

df['Pclass'].unique() #how many unique passenger classes are there?
df['Cabin'].value_counts()

df['Deck'] = df['Cabin'].str[0] #create new column 'Deck' from first letter of 'Cabin'
print(df['Deck'].unique())
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df['Deck'].value_counts()
deck_survival = df.groupby('Deck')['Survived'].mean()*100

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(deck_survival.sort_values(ascending=False))

# Check the deck and class of those who paid $0.00
df[df['Fare'] == 0][['Name', 'Pclass', 'Deck', 'Survived']]

class_survival = df.groupby('Pclass')['Survived'].mean()*100
print(class_survival.sort_values(ascending=False))

deck_price = df.groupby('Deck')['Fare'].mean().sort_values(ascending=False)
print(deck_price)

print(df.groupby('Pclass')['Age'].mean())

df[df['Age'] == 80][['Name', 'Pclass', 'Survived', 'Fare']]
df[df['Age'] == 0.42][['Name', 'Pclass', 'Survived', 'Fare']]
df['Age'] = df['Age'].fillna(df.groupby('Pclass')['Age'].transform('mean')) #impute missing ages with mean age per class
print(df['Age'].isnull().sum())

df.to_csv('../data/titanic_cleaned.csv', index=False)

```
