```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/titanic_cleaned.csv')

sns.set_theme(style='whitegrid') #set the theme for plots

plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class on the Titanic')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')

print(df.groupby('Sex')['Survived'].mean()*100)

plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df, palette='magma')
plt.title('Survival Rate: Class X Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')

# Calculate and print the probability of survival for 3rd class
prob_3rd_female_survival = df[(df['Pclass'] == 3) & (df['Sex'] == 'female')]['Survived'].mean()*100 
print(prob_3rd_female_survival)

# Detailed counts of passengers by class
print(df.value_counts('Sex'))
counts = df.groupby(['Pclass', 'Sex', 'Survived']).size().unstack(fill_value=0)
print(counts)

#deep dive into children survival rates
df['child'] = df['Age'] < 18
child_survival = df[df['child']].groupby('Pclass')['Survived'].mean()*100
print(child_survival)

#Did sex matter for children?
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df[df['child']], palette='magma')
plt.title('Child Survival Rate: Class X Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

#Boxplt of fares by class and survival
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df, palette='pastel')
plt.title('Fare Distribution by Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

# Run your plot code...
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df[df['child']], palette='magma')
plt.title('Child_Survival_Rate:_Class_X_Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Then save it before plt.show()
plt.savefig('../results/Child_Survival_Rate:_Class_X_Gender', dpi=300)
plt.show()

```
